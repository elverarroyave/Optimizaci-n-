$\huge{\bf{Optimizacion}}$

$\bf{Criterio\ de\ la\ primera\ derivada:}$

Ejemplo:
    
$f(x) = x^3 - 12x$

$f'(x) = 3x^2 - 12$

Siguiente paso: $f'(x)=0$

$3x^2 -12 = 0  \Longrightarrow 3(x^2 - 4) = 0 \Longrightarrow  3(x + 2)(x - 2) = 0 $ 

$x_1=-2 \hspace{0.5cm} y \hspace{0.5cm} x_2 = 2$ $\it{Son\ puntos\ criticos}$

$(-\infty,2);(-2,2);(2,+\infty)$
Escogemos un n√∫mero en cada intervalo.

$f'(-3) = 3*(-3)^2-12=27-12=15$ como el resultado es positivo la grafica en este intervalo crece.

$f'(0) = 3*(0)^2-12=-12$ como el resultado es negativo la grafica en este intervalo decrece.

$f'(3) = 3*(3)^2-12=15$ como el resultado es positivo la grafica en este intervalo crece.







```python
##import numpy as np
##import matplotlib.pyplot as plt
from sympy import *
from sympy.abc import x
from sympy.plotting import plot

f = x**3 -12*x

df = diff(f,x) #Primera derivada
d2f = diff(f,x,2) #Segunda derivada

puntos_criticos = solve(Eq(df, 0)) #hallamos los puntos criticos

puntos_infleccion = solve(Eq(d2f,0)) 
```

```python
puntos_criticos
```

```python
puntos_infleccion
```

```python
g=plot(f, (x,-2.5,2.5), line_color='red')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    f = x**3 -12*x
    return f
x=np.linspace(-4,4,100)
plt.plot(x,f(x))
plt.grid()
for x in puntos_criticos:
    plt.plot(x,f(x),'or')
print(puntos_criticos)
```

```python
x=np.linspace(-5,5,100)
plt.plot(x,f(x))
plt.grid()
for x in puntos_criticos:
    plt.plot(x,f(x),'or')
```

$f'(x) = 3x^2+2x-5$

```python
from sympy import *
from sympy.abc import x
from sympy.plotting import plot

f = x**3 + x**2 - 5*x -5

df = diff(f,x) #Primera derivada
d2f = diff(f,x,2) #Segunda derivada

puntos_criticos = solve(Eq(df, 0)) #hallamos los puntos criticos

puntos_infleccion = solve(Eq(d2f,0))  #hallamos los puntos de infleccion
```

```python
puntos_criticos
```
