$\huge{Búsqueda\ de\ la\ razón\ dorada}$

$\large{Por:\ Elver\ Arroyave}$

Hallar el valor máximo de $f(x)=sin(x)-\dfrac{x^2}{10}$ en $x_l=0$ y $x_u=4$

```python
import matplotlib.pyplot as plt
import numpy as np

xL=int(input("Introduce el valor de xL: "))
xU=int(input("Introduce el valor de XU: "))

xInicial = xL
xFinal = xU

d = 0.618*(xU-xL)
x1 = xL + d
x2 = xU - d

def f(x):
    f = 2 * np.sin(x) - (x**2) / 10
    return f

contador = 0

while (abs(x2-x1) > 0.01):
    
    contador = contador + 1
    
    if(f(x2)>f(x1)):
        xU = x1
    else:
        xL = x2

    d = 0.618*(xU - xL)

    x1 = xL + d
    x2 = xU - d

print("El resultado es: ",x1 , "El numero de iteraciones es:", contador)

```

```python
x=np.linspace(xInicial,xFinal,100)
plt.plot(x,f(x))
plt.grid()
plt.plot(x1,f(x1),'or')
```

```python
x=np.linspace(-10,10,100)
plt.plot(x,f(x))
plt.grid()
plt.plot(x1,f(x1),'or')
```
